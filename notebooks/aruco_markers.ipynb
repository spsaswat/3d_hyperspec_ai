```python
import os
import cv2

# Output folder
output_dir = "aruco_markers"
os.makedirs(output_dir, exist_ok=True)

# Choose dictionary
aruco_dict = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_4X4_50)

# Marker settings
marker_ids = [0, 1, 2, 3]
marker_size_px = 600  # image size in pixels

# Corner naming
corner_names = {
    0: "top_left",
    1: "top_right",
    2: "bottom_left",
    3: "bottom_right",
}

for marker_id in marker_ids:
    marker_img = cv2.aruco.generateImageMarker(aruco_dict, marker_id, marker_size_px)

    filename = f"aruco_{marker_id}_{corner_names[marker_id]}.png"
    save_path = os.path.join(output_dir, filename)

    cv2.imwrite(save_path, marker_img)
    print(f"Saved: {save_path}")

print("Done.")
```
